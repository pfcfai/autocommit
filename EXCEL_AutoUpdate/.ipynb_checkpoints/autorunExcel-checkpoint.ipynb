```python
# Import packages
import os
import win32com.client

# Create Excel Application
xlApp=win32com.client.Dispatch("Excel.Application")

# In default, Python open excel book visibly.
xlBook=xlApp.Workbooks.Open(Filename='/home/spark/autocommit/EXCEL_AutoUpdate/test.xlsm')
#< <COMObject Open>
xlApp.DisplayAlerts = False #把警示關掉爽拉

# Same as R code
xlApp.Application.Run("按鈕1_Click")



# 保存并关闭工作簿
xlBook.SaveAs('/home/spark/autocommit/EXCEL_AutoUpdate/test.xlsm')
xlBook.Close()
# 关闭应用
xlApp.Quit()
```
